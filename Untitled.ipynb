```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori  
import seaborn as sns
```

```python
data = pd.read_csv('google_review_ratings.csv')  
```

```python
data=data.drop('Unnamed: 25',axis=1)
```

```python
data1=data.drop('User',axis=1)
```

```python
fig, ax = plt.subplots(figsize=(50,50))
sns.heatmap(data.corr(), vmax=0.6, square=True, annot=True)
```

```python
print(data1.dtypes)
cols = data1.select_dtypes(exclude=['float']).columns

data1[cols] = data1[cols].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
data1=data1.fillna(0)
```

```python
from sklearn import cluster, datasets, metrics


kmeans_fit = cluster.KMeans(n_clusters = 5).fit(data1)

# 印出分群結果
cluster_labels = kmeans_fit.labels_
print("分群結果：")
print(cluster_labels)
print("---")
```

```python

silhouette_avg = metrics.silhouette_score(data1, cluster_labels)
print(silhouette_avg)
```

```python
cluster_labels=pd.Series(cluster_labels)
data = pd.concat([data1, cluster_labels.rename('cluster')], axis=1)
```

```python
c0=data.loc[data['cluster'] == 0]
c0=c0.apply(np.mean)
```

```python
c1=data.loc[data['cluster'] == 1]
c1=c1.apply(np.mean)
```

```python
c2=data.loc[data['cluster'] == 2]
c2=c2.apply(np.mean)
```

```python
c3=data.loc[data['cluster'] == 3]
c3=c3.apply(np.mean)
```

```python
c4=data.loc[data['cluster'] == 4]
c4=c4.apply(np.mean)
```

```python
c0=pd.Series(c0)
c1=pd.Series(c1)
c2=pd.Series(c2)
c3=pd.Series(c3)
c4=pd.Series(c4)
```

```python
conclusion = pd.DataFrame()
conclusion = pd.concat([conclusion, c0.rename('c0')], axis=1,sort=False)
conclusion = pd.concat([conclusion, c1.rename('c1')], axis=1,sort=False)
conclusion = pd.concat([conclusion, c2.rename('c2')], axis=1,sort=False)
conclusion = pd.concat([conclusion, c3.rename('c3')], axis=1,sort=False)
conclusion = pd.concat([conclusion, c4.rename('c4')], axis=1,sort=False)
```
